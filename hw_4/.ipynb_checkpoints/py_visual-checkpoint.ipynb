```python
import numpy as np

my_data = np.array([np.loadtxt("numbers1.txt").T, np.loadtxt("numbers2.txt").T, np.loadtxt("numbers3.txt").T, np.loadtxt("numbers4.txt").T, np.loadtxt("numbers5.txt").T])

N = 10000
k = 5
r = 10

x = np.zeros((k, N))
v = np.zeros((k, N))
t = np.zeros((k, N))

for i in range(k):
    x[i] = my_data[i][0]
    v[i] = my_data[i][1]
    t[i] = my_data[i][2]

```

```python
import matplotlib.pyplot as plt



E = np.zeros((k, N))
for i in range(k):
    E[i] = 0.5 * ((x[i])**2 + (v[i])**2)



```

```python
import matplotlib.pyplot as plt

plot_num = int(input("Введите количество графиков, которые Вы ходите построить, одним числом: "))




plt.title("E(t)")
for i in range(plot_num):
    plt.plot(t[i], E[i])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plot_num = int(input("Введите количество графиков, которые Вы ходите построить, одним числом: "))

plt.title("v(x)")
for i in range(plot_num):
    plt.plot(x[i], v[i])


theta = np.linspace(0, 2 * np.pi, 100)
x_teor = r * np.cos(theta)
v_teor = r * np.sin(theta)
    
plt.plot(x_teor, v_teor)
    
plt.show()
```

```python
#Predictor–corrector method

import numpy as np

pc_data = np.array([np.loadtxt("pc_numbers1.txt").T, np.loadtxt("pc_numbers2.txt").T, np.loadtxt("pc_numbers3.txt").T, np.loadtxt("pc_numbers4.txt").T, np.loadtxt("pc_numbers5.txt").T])

N = 10000
k = 5
r = 10

pc_x = np.zeros((k, N))
pc_v = np.zeros((k, N))
pc_t = np.zeros((k, N))

for i in range(k):
    pc_x[i] = pc_data[i][0]
    pc_v[i] = pc_data[i][1]
    pc_t[i] = pc_data[i][2]


```

```python
import matplotlib.pyplot as plt
import numpy as np

plot_num = int(input("Введите номер от 1 до 5 одним числом, который отвечает за величину дискретизации времени, график для которого вы хотите построить: "))

plt.title("v(x) and pc_v(pc_x)")

plt.plot(x[plot_num-1], v[plot_num-1], 'r')


plt.plot(pc_x[plot_num-1], pc_v[plot_num-1], 'b')



    
plt.show()
```
