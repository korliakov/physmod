```python
import matplotlib.pyplot as plt
import numpy as np

def try_ex_func():
    user_input = input("Введите 4 отношения частот, фигуры Лиссажу для которых Вы хотите получить (например 1:1, 2:1, 4:3, 5:7): ")
    try:
        if (user_input.find(':') == -1):
            print("Неверный ввод.")
            return try_ex_func()

        frequencies = [(int(el.split(':')[0]), int(el.split(':')[1])) for el in user_input.strip().split(', ')]   #Питон могет)
    except Exception:
        print("Неверный ввод.")
        return try_ex_func()
        
    else:
        print("Ниже приведены 4 фигуры Лиссажу.")
        return frequencies

freq = try_ex_func()


fig, ax = plt.subplots(2, 2)

A = np.linspace(0, 6*np.pi, 1000)

ax[0, 0].plot(np.cos(freq[0][0]*A), np.sin(freq[0][1]*A), color = 'black')
ax[0, 0].scatter(np.cos(freq[0][0]*A), np.sin(freq[0][1]*A), color = 'red', s = 5)

ax[1, 0].plot(np.cos(freq[1][0]*A), np.sin(freq[1][1]*A), color = 'black')
ax[1, 0].scatter(np.cos(freq[1][0]*A), np.sin(freq[1][1]*A), color = 'red', s = 5)

ax[0, 1].plot(np.cos(freq[2][0]*A), np.sin(freq[2][1]*A), color = 'black')
ax[0, 1].scatter(np.cos(freq[2][0]*A), np.sin(freq[2][1]*A), color = 'red', s = 5)

ax[1, 1].plot(np.cos(freq[3][0]*A), np.sin(freq[3][1]*A), color = 'black')
ax[1, 1].scatter(np.cos(freq[3][0]*A), np.sin(freq[3][1]*A), color = 'red', s = 5)


```

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.array([np.cos(freq[1][0]*A), np.sin(freq[1][1]*A)])
a_squared = np.multiply(a,a)
 
dist = a_squared[0] + a_squared[1] #удаленность каждой точки от центра координат в квадрате

plt.title('Гистограма для последнего введенного соотношения частот')
plt.hist(dist, 30, density=False)
```

```python
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(0, 10, 10) + np.random.normal(0,1,10) #добавил шумов
Y = np.linspace(0, 10, 10) + np.random.normal(0,1,10) #добавил шумов

A = np.vstack([X, np.ones(len(X))]).T

m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

import matplotlib.pyplot as plt
plt.plot(X, Y, 'o', label='Original data', markersize=10)
plt.plot(X, m*X + c, 'r', label='Fitted line')
plt.legend()
plt.show()

print(m,c)
```

```python
import datetime

def wrapper(func):
    def inner():
        time_prev = datetime.datetime.now()
        print(f'Начало: {time_prev}')
        func()
        time_delta = datetime.datetime.now()-time_prev
        print(f'Конец: {datetime.datetime.now()}')
        print(f'Время работы: {time_delta}')
    return inner
        
        
        
        
```

```python
import numpy as np

a = np.zeros(10000000, dtype=np.float)
a[0] = 10

b = np.zeros(10000000, dtype=np.float)

delta = 0.00005



@wrapper
def func():
    i = 0
    while (i < len(a)-1):
        b[i+1] = b[i] - delta * a[i]
        a[i+1] = a[i] + delta * b[i]
        i+=1
    

func()

```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.plot(a, b)
```
