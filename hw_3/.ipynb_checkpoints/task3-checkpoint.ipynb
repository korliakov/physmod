```python
import datetime

def wrapper(func):
    def inner(*args):
        time_prev = datetime.datetime.now()
        print(f'Начало: {time_prev}')
        res = func(*args)
        time_delta = datetime.datetime.now()-time_prev
        print(f'Конец: {datetime.datetime.now()}')
        print(f'Время работы: {time_delta}')
        return res 
    return inner
```

```python
@wrapper
def series_sum(base_degree: int):
    sum = 0
    for i in range (1, 100):
        sum+=(base_degree)**(i)
    return sum
    
```

```python
print(series_sum(0.5))
```

```python
#simple problem

import numpy as np

a = np.zeros(1000, dtype=np.float)
a[0] = 10

t = np.zeros(1000, dtype=np.float)
t[0] = 0

b = np.zeros(1000, dtype=np.float)

delta = 0.05



@wrapper
def func():
    i = 0
    while (i < len(a)-1):
        b[i+1] = b[i] - delta * a[i]
        a[i+1] = a[i] + delta * b[i]
        i+=1
        t[i+1] = t[i] + delta
    

func()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.plot(a, b)
```
