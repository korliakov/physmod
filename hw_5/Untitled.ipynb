```python
import numpy as np

data = np.loadtxt("numbers.txt").T

x = data[0]
v = data[1]
t = data[2]

```

```python
import matplotlib.pyplot as plt

plt.title("x(v)")
plt.plot(v, x)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.title("x(t)")
plt.plot(t, x)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

E = (x*x + v*v)/2

plt.title("E(t)")
plt.plot(t, E)
plt.show()
```
